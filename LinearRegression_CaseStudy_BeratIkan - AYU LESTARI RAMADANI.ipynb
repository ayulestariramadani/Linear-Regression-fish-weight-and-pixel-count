# Linear Regression Case Study

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("ikan.csv")
df.head()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
plt.scatter(train.jumlahpixel, train.berat,  color='red')
plt.xlabel("jumlahpixel")
plt.ylabel("berat")
plt.show()
```

# Menggunakan Library pada Python (Scikit-Learn)

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['jumlahpixel']])
train_y = np.asanyarray(train[['berat']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
linearRegressionInPython = 'Persamaan Garis yang diperoleh : y = '+ str(regr.coef_[0][0]) + ' x ' + str(regr.intercept_[0])
```

```python
plt.scatter(train.jumlahpixel, train.berat,  color='red')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Jumlah Pixel")
plt.ylabel("Berat")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['jumlahpixel']])
test_y = np.asanyarray(test[['berat']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```

# Bangun model menggunakan rumus Linear Regression


```python
# Menghitung Rata - Rata x dan y

mean_x = train_x.mean()
mean_y = train_y.mean()
print('Rata - rata x : ', mean_x)
print('Rata - rata y : ', mean_y)
```

```python
pembilang = ((train_x - mean_x) * (train_y - mean_y)).sum()
penyebut = ((train_x - mean_x)**2).sum()
b = pembilang / penyebut
b
```

```python
a = mean_y - b * mean_x
a
```

```python
linearRegression = 'Persamaan Garis yang diperoleh : y = ' + str(b) + ' x ' + str(a)
print(linearRegression)
```

```python
## Plot Model
plt.scatter(train_x, train_y,  color='green')
plt.plot(train_x, b*train_x + a, '-r')
plt.xlabel("Jumlah Pixel")
plt.ylabel("Berat")
```
